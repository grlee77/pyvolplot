```python
%matplotlib inline
import numpy as np

from pyvolplot import volshow
```

## Import a color dataset (magnitude of first eigenvector for DTI)

```python
data=np.load('DTI_color_V1abs.npz')['DTI_v1']
```

Create a montage of every 4th slice

```python
ax = volshow(data[:,-1::-1,::4],mode='m',isRGB=True) #,show_lines=True)
```

```python
ax = volshow(data[:,-1::-1,-1::-1],mode='c',isRGB=True,notick=True)
```

Both views can be combined into a single plot in a few different ways.  One option is to pass in an existing figure and specify the subplots to use:

```python
from matplotlib import pyplot as plt
f = plt.figure(figsize=[16, 16])
volshow(data[:,-1::-1,-1::-1], fig=f, subplot=121, mode='c', isRGB=True, notick=True)
volshow(data[:,-1::-1,::4], fig=f, subplot=122, mode='m', isRGB=True)
```

Alternatively, the figure and subplots can be pre-generated and the axes passed in:
    

```python
from matplotlib import pyplot as plt
f, (ax1, ax2) = plt.subplots(1,2)
volshow(data[:,-1::-1,-1::-1], ax=ax1, mode='c', isRGB=True, notick=True)
volshow(data[:,-1::-1,::4], ax=ax2, subplot=122, mode='m', isRGB=True, figsize=[16, 8])
```

Finally, the first argument to volshow and many of the keywords support list input

```python
from matplotlib import pyplot as plt
vol_list = [data[:,-1::-1,-1::-1],
            data[:,-1::-1,::4]]
im_list = volshow(vol_list, mode=['c', 'm'],
                  isRGB = [True, True], notick=[True, False], figsize=[16, 8],
                  facecolor='k', edgecolor='w')
```
